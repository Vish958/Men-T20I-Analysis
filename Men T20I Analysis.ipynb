   # *Exploring and Visualizing the data

```python

df['Innings Player'].value_counts().plot(kind = 'bar')
plt.ylabel('Runs' , color = 'red' , size = 'x-large')
plt.xlabel(' Players Name' , color = 'red' , size = 'x-large')
```

```python
top12 = df['Innings Player'].value_counts()
top12.sort_values(ascending=False)
top12 = top12[:12]
top12.plot(kind='bar')
plt.ylabel('Runs' , color='red' , size = 'x-large')
plt.xlabel('Players Name' , color='red' , size = 'x-large')
plt.show()
```

```python
top20 = df['Innings Player'].value_counts()
top20.sort_values(ascending=False)
top20 = top20[:20]
top20.plot(kind='bar')
plt.ylabel('Runs' , color='red' , size = 'x-large')
plt.xlabel('Players Name' , color='red' , size = 'x-large')
plt.show()
```

```python
top12 = df['Innings Player'].value_counts()
top12.sort_values(ascending=False)
top12 = top12[:12]
top12.plot(kind='pie')
plt.show()
```

```python
top20 = df['Innings Player'].value_counts()
top20.sort_values(ascending=False)
top20 = top20[:20]
top20.plot(kind = 'pie')
plt.show()
```

```python
labels = 'RG Sharma','MS Dhoni','MJ Guptil', 'V Kohli'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### *I am reveal patterns in our data by using some math operations, such as sum and count. By using the value_counts function, we can count unique values in a column. Then, we can call the plot function on the result to create a bar chart.

```python
df['Innings Boundary Fours'].value_counts().plot(kind = 'bar')
```

```python
top10 = df['Innings Boundary Fours'].value_counts()
top10.sort_values(ascending=False)
top10 = top10[:10]
top10.plot(kind = 'pie')
plt.show()
```

```python
labels = '1' , '2' , '3' ,'4' '10'
sizes = [15, 30, 45, 60]
explode = (0, 0.2, 0, 0)  # only "explode" the 2nd slice (i.e. 'Fours')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
